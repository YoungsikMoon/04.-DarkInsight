```python
# yolov7/cgf/training/custom-yolov7-tiny.yaml --> 가장 윗 줄 클레스 수 바꿔줌
import shutil

shutil.copy("/content/yolov7/cfg/training/yolov7-tiny.yaml", "/content/yolov7/cfg/training/custom-yolov7-tiny.yaml")
```

```python
# 증강 이미지 모두 한 곳으로 합쳐주기

# train 증강파일
%cd /content/
from glob import glob

train_aug_file = glob("/content/train_resized/aug_images/*")
train_aug_lable = glob("/content/train_resized/aug_labels/*")
for image, label in zip(train_aug_file, train_aug_lable):
    shutil.move(image, "/content/train_resized/images")
    shutil.move(label, "/content/train_resized/labels")

# valid 증강파일
valid_aug_file = glob("/content/valid_resized/aug_images/*")
valid_aug_lable = glob("/content/valid_resized/aug_labels/*")
for image, label in zip(valid_aug_file, valid_aug_lable):
    shutil.move(image, "/content/valid_resized/images")
    shutil.move(label, "/content/valid_resized/labels")

# test 증강파일
test_aug_file = glob("/content/test_resized/aug_images/*")
test_aug_lable = glob("/content/test_resized/aug_labels/*")
for image, label in zip(test_aug_file, test_aug_lable):
    shutil.move(image, "/content/test_resized/images")
    shutil.move(label, "/content/test_resized/labels")

len(glob("/content/train_resized/images/*")), len(glob("/content/train_resized/labels/*")), len(glob("/content/valid_resized/images/*")), len(glob("/content/valid_resized/labels/*")), len(glob("/content/test_resized/images/*")), len(glob("/content/test_resized/labels/*"))
```
