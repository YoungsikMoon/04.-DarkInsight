```python
import json
import matplotlib.pyplot as plt
from numpy import sqrt

falldown_s_path = r'.\test_video\tracking\v5_small_head_64_30_falldown_small.json'

with open(falldown_s_path, 'r') as f:
    result_fs = json.load(f)
route = result_fs["ps_all_route"]["0"]
f = [x[0] for x in route]
x = [x[1] for x in route]
y = [x[2] for x in route]
w = [x[3] for x in route]


dist = [sqrt((x[i]-x[i-5])**2 + (y[i]-y[i-5])**2) for i in range(len(f)) if i > 4]
dist_norm = [(sqrt((x[i]-x[i-5])**2 + (y[i]-y[i-5])**2))/w[i] for i in range(len(f)) if i > 4]

fig = plt.figure(figsize=(15,5))

# falldown-small(5Frame), B-Nomalization
fig.add_subplot(1, 2, 1)
plt.plot(f[5:], dist)

plt.axvline(25, 0, 150, color='red', linestyle='--', linewidth=2)
plt.axvline(55, 0, 150, color='red', linestyle='--', linewidth=2)

plt.axhline(dist[25-5], 0, 130, color='red', linestyle='--', linewidth=2)
plt.axhline(max(dist), 0, 130, color='red', linestyle='--', linewidth=2)
print(dist[25-5], max(dist))

plt.title("falldown-small(5Frame), B-Nomalization")

# falldown-small(5Frame), A-Nomalization
fig.add_subplot(1, 2, 2)
plt.plot(f[5:], dist_norm)

plt.axvline(25, 0, 150, color='red', linestyle='--', linewidth=2)
plt.axvline(55, 0, 150, color='red', linestyle='--', linewidth=2)

plt.axhline(dist_norm[25-5], 0, 130, color='red', linestyle='--', linewidth=2)
plt.axhline(max(dist_norm), 0, 130, color='red', linestyle='--', linewidth=2)
print(dist_norm[25-5], max(dist_norm))

plt.title("falldown-small(5Frame), A-Nomalization")

plt.show()
```
